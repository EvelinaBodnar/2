**Метод Хука — Дживса** або пошук за зразком так само, як і метод Нелдера–Міда, призначений для пошуку безумовного локального екстремуму функції і відноситься до прямих методів, тобто спирається безпосередньо на значення функції. Алгоритм складається з двох фаз: досліджуючий пошук і пошук за зразком.

Переваги

Даний алгоритм володіє властивістю «прискорюватися», що сприяє підвищенню його загальної ефективності. Друга перевага методу Хука—Дживса — можливість отримання за його допомогою наближеного рішення, якість якого безупинно підвищується на всіх стадіях чисельного рішення. Особливо явно переваги подібних засобів виявляються при відшуканні екстремумів на гіперповерхнях, які містять глибокі вузькі западини, тобто тоді, коли градієнтні методи неефективні. Також перевагами даного методу є проста стратегія пошуку і невеликий обсяг необхідної пам'яті.

Недоліки

Алгоритм заснований на циклічному русі по координатам, що може привести до виродження алгоритму в нескінченну послідовність пошуків через дослідження без пошуку за зразком. При використанні методу передбачається, що цільова функція яка досліджується, є унімодальною, тобто має на інтервалі дослідження тільки один оптимум. Це означає, що розв'язком може стати локальний мінімум, а не глобальний, якщо на досліджуваній області присутні кілька мінімумів.

# Реалізація методу:
Іморт бібліотек;

Вид функції Розенброка;

Метод Хука-Дживса (безумовна оптимізація).

```python
from itertools import groupby
from math import sqrt
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import axes3d

f_calls = 0

def rosenbrock(x, y):
    global f_calls
    f_calls += 1
    return (1-x)**2 + 100*(y-x**2)**2


def hook_jeeves(x0, d, eps, method, a):
    def func(x1, x2):
        return rosenbrock(x1, x2) 
    dx1 = d[0]
    dx2 = d[1]
    x1 = [x0]
    x2 = x0
    y1 = x0
    limiter = True
    while (sqrt(dx1**2 + dx2**2) > eps) if method == 1 else (limiter):
        f_old = func(x2[0], x2[1])
        if func(y1[0] + dx1, y1[1]) < func(y1[0], y1[1]):
            y2 =[y1[0] + dx1, y1[1]]
        elif func(y1[0] - dx1, y1[1]) < func(y1[0], y1[1]):
            y2 = [y1[0] - dx1, y1[1]]
        else:
            y2 = y1
        if func(y2[0], y2[1] + dx2) < func(y2[0], y2[1]):
            y3 =[y2[0], y2[1] + dx2]
        elif func(y2[0], y2[1] - dx2) < func(y2[0], y2[1]):
            y3 = [y2[0], y2[1] - dx2]
        else:
            y3 = y2

        f_y3 = func(y3[0], y3[1])
        if f_y3 < f_old:
            x2 = y3
            y1 = [2*x2[0] - x1[-1][0], 2*x2[1] - x1[-1][1]]
            x1.append(x2)
        else:
            dx1 /= a
            dx2 /= a
            x1.append(x1[-1])
            x2 = x1[-1]
            y1 = x2
        if method != 1:
            uniq = [el for el, _ in groupby(x1)]
            if len(uniq) > 1:
                limiter = sqrt((uniq[-1][0] - uniq[-2][0])**2 +
                               (uniq[-1][1] - uniq[-2][1])**2)/sqrt(uniq[-2][0]**2 + uniq[-2][1]**2) > eps\
                          or (abs(func(uniq[-1][0], uniq[-1][1]) - func(uniq[-2][0], uniq[-2][1]))) > eps

            else:
                break
    return [round(x2[0], 7), round(x2[1], 7), len(x1), x1]


result = hook_jeeves([1.5, -0.5], [0.1, 0.1], 10**(-5), 1, 2)
xmin = min([result[3][i][j] for i in range(len(result[3])) for j in range(len(result[3][i]))])
xmax = max([result[3][i][j] for i in range(len(result[3])) for j in range(len(result[3][i]))])

ax = axes3d.Axes3D(plt.figure())
i = np.arange(xmin-1, xmax+1, 0.01)
X, Y = np.meshgrid(i, i)
Z = rosenbrock(X, Y)
ax.plot_wireframe(X, Y, Z, rstride=100, cstride=100)

print('Кінцева Х за методом Хука-Дживса: ', result[0], result[1])
print('Кількість вдалих точок: ', result[2])
print('Кількість обчислень функції: ',f_calls)

ax.scatter([r[0] for r in result[3]],
           [r[1] for r in result[3]],
           [rosenbrock(result[3][i][0],result[3][i][1]) for i in range(len(result[3]))], s=10, color='red')

ax.plot3D([r[0] for r in result[3]],
           [r[1] for r in result[3]],
           [rosenbrock(result[3][i][0],result[3][i][1]) for i in range(len(result[3]))], color='orange')

plt.show()

```

Умовна оптимізація з лінійним обмеженням:

Значення параметра в алгоритмі Свена;

Метод одновимірного пошуку ДСК-Пауела та Золотого пертину;



```python
LINEAR = 1


def rosenbrock(x, y):

    if x+y > LINEAR:
        return 9999999
    else:
        return rs(x, y)

def rs(x,y):
    return (1-x)**2 + 100*(y-x**2)**2
 
xmin = min([result[3][i][j] for i in range(len(result[3])) for j in range(len(result[3][i]))])
xmax = max([result[3][i][j] for i in range(len(result[3])) for j in range(len(result[3][i]))])
result = hook_jeeves([-0.5, 0], [0.1, 0.1], 10**(-3), 1, 2)
x1 = (0, LINEAR)
x2 = (0, LINEAR)
xmin = min([result[3][i][j] for i in range(len(result[3])) for j in range(len(result[3][i]))])
xmax = max([result[3][i][j] for i in range(len(result[3])) for j in range(len(result[3][i]))])
xmin = min([xmin, x1[0], x1[1], x2[0], x2[1]])
xmax = max([xmax, x1[0], x1[1], x2[0], x2[1]])
ax = axes3d.Axes3D(plt.figure())
i = np.arange(xmin-1, xmax+1, 0.01)
X, Y = np.meshgrid(i, i)
Z = rs(X, Y)
ax.plot_wireframe(X, Y, Z, rstride=100, cstride=100)

print('Кінцева Х за методом Хука-Дживса: ', result[0], result[1])
print('Кількість вдалих точок: ', result[2])
print('Кількість обчислень функції: ',f_calls)

z = [rs(result[3][i][0],result[3][i][1]) for i in range(len(result[3]))]
ax.scatter([r[0] for r in result[3]], [r[1] for r in result[3]], z, s=10, color='red')


ax.plot3D([r[0] for r in result[3]],
           [r[1] for r in result[3]],
           [rosenbrock(result[3][i][0],result[3][i][1]) for i in range(len(result[3]))], color='orange')

plt.show()


S = [0, 0]

if rs(result[0] + 0.1, result[1] - 0.1) > rs(result[0] - 0.1, result[1] + 0.1):
    S = [LINEAR, -LINEAR]
else:
    S = [-LINEAR, LINEAR]


def rs_m(lm):
    x1 = result[0] - lm*S[0]
    x2 = result[1] - lm*S[1]
    return rs(x1, x2)

def func(x):
    return rs_m(x)


def Sven(x, dx):
    def f(a):
        return func(a)
    la = 0
    if f(x + dx) <= f(x):
        la = 1
    elif f(x - dx) <= f(x):
        la = -1
    xlist = []
    end = False
    while not end:
        xlist.append(round(x, 5))
        fx = f(x)
        x += la*dx
        if f(x) < fx:
            dx *= 2
        else:
            dx /= 2
            x -= la*dx
            xlist.append(round(x, 5))
            end = True

    flist = [f(xlist[i]) for i in range(len(xlist))]
    minf = min(flist)
    minf_indx = flist.index(minf)
    print('Результат алгоритма Свена: ', xlist[minf_indx-1], xlist[minf_indx+1])
    return xlist[minf_indx-1], xlist[minf_indx+1]


def DSKP(a, b, eps):
    def f(x):
        return func(x)
    xlist = [a, (a+b)/2, b]
    flist = [f(x) for x in xlist]
    if not (flist[0] >= flist[1] <= flist[2]):
        if flist[0] < flist[2]:
            flist[0], flist[1], flist[2] = flist[1], flist[0], flist[2]
            xlist[0], xlist[1], xlist[2] = xlist[1], xlist[0], xlist[2]
        else:
            flist[0], flist[1], flist[2] = flist[0], flist[2], flist[1]
            xlist[0], xlist[1], xlist[2] = xlist[0], xlist[2], xlist[1]
    end = False
    xs = xlist[1]
    while not end:
        if round(abs(xlist[0] - xlist[1]), 4) == round(abs(xlist[1] - xlist[2]), 4):
            xs = xlist[1] + (abs(xlist[0] - xlist[1])*(flist[0] - flist[2]))/(2*(flist[0] - 2*flist[1] + flist[2]))
        else:
            a1 = (flist[1] - flist[0])/(xlist[1] - xlist[0])
            a2 = (((flist[2] - flist[0])/(xlist[2] - xlist[0])) - a1)/(xlist[2] - xlist[1])
            xs = (xlist[0] + xlist[1])/2 - a1/(2*a2)
        if abs(xlist[1] - xs) < eps and abs(flist[1] - f(xs)) < eps:
            end = True
        else:
            xlist.append(xs)
            flist.append(f(xs))
            maxf = max(flist)
            maxf_indx = flist.index(maxf)
            del xlist[maxf_indx]
            xlist.sort()
            flist = [f(x) for x in xlist]
    print('Список точок х: ', xlist)        
    return round(xs, 5)

def GS(a, b, eps):
    def f(x):
        return func(x)
    end = False
    if a > b:
        a, b = b, a
    while not end:
        L = abs(b-a)
        x1 = a + 0.382*L
        x2 = a + 0.618*L
        if f(a) > f(b):
            a = x1
        else:
            b = x2
        if abs(b-a) <= eps:
            end = True
    print('Критерій зупинки досягнуто: ', abs(b-a))
    return round((b+a)/2, 5)


l = Sven(1, 0.01)
D = (DSKP(l[0], l[1], 10**(-5)))
G = (GS(l[0], l[1], 10**(-5)))

print(result[0] - D*S[0], result[1] - D*S[1])
print(result[0] - G*S[0], result[1] - G*S[1])
```
